```python
imoveis = dados_brutos.copy()
```

```python
imoveis['Bairros'] = imoveis['Bairros'].apply(lambda x: x.split(',')[1].split('\n')[0].strip() if ',' in x else x)
imoveis['Bairros']
```

```python
imoveis = imoveis[imoveis['Bairros'] != 'Blumenau\nCód.: 6301173']
```

```python
# Remover as linhas onde a coluna 'Valores' é 'CONSULTAR'
imoveis = imoveis[imoveis['Valores'] != 'CONSULTAR']

# Agora, converter os valores numéricos corretamente
imoveis['Valores'] = imoveis['Valores'].apply(lambda x: int(x.replace('R$', '').replace('.', '').replace(',', '').strip()))

imoveis
```

```python
imoveis['Valores'] = (imoveis['Valores'] / 100).round(0)
imoveis
```

```python
# Substituir '\n' por '|' na coluna 'Informações'
imoveis['Informações'] = imoveis['Informações'].replace('\n', ' | ', regex=True)

imoveis
```

```python
# Função para extrair os dados das novas colunas, retornando 0 quando não encontrar valores
def extrair_informacoes(info):
    if isinstance(info, str):  # Verifica se o valor é uma string
        # Separar as informações por '|'
        partes = info.split(' | ')
        
        # Extrair os valores para cada coluna
        quartos = next((p for p in partes if 'quarto' in p.lower()), '0')
        suites = next((p for p in partes if 'suíte' in p.lower()), '0')
        vagas = next((p for p in partes if 'vaga' in p.lower()), '0')
        tamanho = next((p for p in partes if 'm²' in p.lower()), '0')
        
        # Retornar os valores limpos
        return [quartos, suites, vagas, tamanho]
    else:
        # Retornar 0 para casos não string
        return ['0', '0', '0', '0']

# Aplicar a função à coluna 'Informações'
imoveis[['Quartos', 'Suítes', 'Vagas', 'Tamanho']] = imoveis['Informações'].apply(extrair_informacoes).apply(pd.Series)

# Limpar os valores de cada coluna (remover textos e converter para numérico)
imoveis['Quartos'] = imoveis['Quartos'].str.extract('(\d+)').astype(float).fillna(0)
imoveis['Suítes'] = imoveis['Suítes'].str.extract('(\d+)').astype(float).fillna(0)
imoveis['Vagas'] = imoveis['Vagas'].str.extract('(\d+)').astype(float).fillna(0)
imoveis['Tamanho'] = imoveis['Tamanho'].str.extract('(\d+,\d+|\d+)').replace(',', '.', regex=True).astype(float).fillna(0)

# Exibir o DataFrame com as novas colunas
imoveis[['Quartos', 'Suítes', 'Vagas', 'Tamanho']]

```

```python
# Função para classificar a descrição como 'Casa', 'Apartamento' ou 'Outro'
def classificar_tipo(descricao):
    if isinstance(descricao, str):
        if 'Casa' in descricao:
            return 'Casa'
        elif 'Apartamento' in descricao:
            return 'Apartamento'
    return 'Outro'  # Retorna 'Outro' se não for uma string ou não houver correspondência

# Aplicar a função à coluna 'Descricao'
imoveis['Tipo'] = imoveis['Descricao'].apply(classificar_tipo)

# Exibir o DataFrame com a nova coluna
imoveis

```

```python
imoveis = imoveis.drop(columns=['Informações', 'Descricao'])
imoveis
```
