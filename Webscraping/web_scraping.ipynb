```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python
url = 'https://www.imoveis-sc.com.br/blumenau/comprar/apartamento+casa'
driver.get(url)
```

```python
driver.find_element(By.CLASS_NAME, 'imovel-extra').text
```

```python
driver.find_element(By.CLASS_NAME, 'imovel-preco').text
```

```python
driver.find_element(By.CLASS_NAME, 'imovel-info').text
```

```python
driver.find_element(By.CLASS_NAME, 'imovel-titulo').text
```

```python
bairros = []
valores = []
infos = []
descricoes = []

while True:
    # Coletar dados da página atual
    bairros.extend([bairro.text for bairro in driver.find_elements(By.CLASS_NAME, 'imovel-extra')])
    valores.extend([valor.text for valor in driver.find_elements(By.CLASS_NAME, 'imovel-preco')])
    infos.extend([info.text for info in driver.find_elements(By.CLASS_NAME, 'imovel-info')])
    descricoes.extend([descricao.text for descricao in driver.find_elements(By.CLASS_NAME, 'imovel-titulo')])
    
    try:
        # Tentar encontrar o botão de "Próxima página" e clicar
        next_button = driver.find_element(By.CLASS_NAME, 'next')  # Substitua pela classe do botão
        next_button.click()

        # Aguarde o carregamento da próxima página (use sleep ou WebDriverWait se necessário)
        time.sleep(5)  # ou use WebDriverWait para uma espera mais eficiente
    except:
        # Se não encontrar o botão ou não puder avançar, quebrar o loop
        break
```

```python
infos
```

```python
df = pd.DataFrame({
    'Bairros': bairros,
    'Valores': valores,
    'Informações': infos,
    'Descricao': descricoes
})

# Salvar o DataFrame em um arquivo CSV
df.to_csv('imoveis.csv', index=False,sep=';', encoding='utf-8-sig')
```
